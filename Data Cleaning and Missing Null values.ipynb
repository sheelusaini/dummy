```python
pd.concat([df['player_of_match'].value_counts()/df.shape[0]*100,
df3_drop_rows['player_of_match'].value_counts()/df3_drop_rows.shape[0]*100],axis=1,
          keys=['player_of_match_org','player_of_match_clean'])
```

```python
def cat_var_dist(var):
    return pd.concat([df[var].value_counts()/df.shape[0]*100,
           df3_drop_rows[var].value_counts()/df3_drop_rows.shape[0]*100],axis=1,
          keys=[var+'_org',var+'_clean'])
```
